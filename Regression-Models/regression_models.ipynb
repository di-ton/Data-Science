```python
# since there are negative distances, we prefer to take their absolute values
## absolute error

def compute_distances(x, y, a, b):
    predictions = a * x + b # y_tilda
    distances = np.abs(y - predictions)
    return distances
```

```python
def compute_total_distance(x, y, a, b):
    predictions = a * x + b # y_tilda
    distances = np.abs(y - predictions)
    total_distance = np.sum(distances)
    return total_distance

## this distance does not work well, since "the more points, the bigger value of the total distance"
```

```python
# mean absolute error

def compute_mean_distance(x, y, a, b):
    predictions = a * x + b # y_tilda
    distances = np.abs(y - predictions)
    total_distance = np.mean(distances)

    return total_distance
```

```python
compute_mean_distance(x, y, -2, 1), compute_mean_distance(x, y, 5, 0), compute_mean_distance(x, y, 2, 1), compute_mean_distance(x, y, 2, 3)
```

```python
# mean square error (MSE) 
## cost_function

def compute_square_distance(x, y, a, b):
    predictions = a * x + b # y_tilda
    distances = (y - predictions) ** 2
    total_distance = np.mean(distances)

    return total_distance
```

```python
compute_square_distance(x, y, -2, 1), compute_square_distance(x, y, 5, 0), compute_square_distance(x, y, 2, 1), compute_square_distance(x, y, 2, 3)
```

```python
# the same function from above, but another name

def compute_cost(x, y, a, b):
    predictions = a * x + b # y_tilda
    distances = (y - predictions) ** 2
    total_distance = np.mean(distances)

    return total_distance
```

```python
y = 3 * x - 2
def compute_gradient(x, y, a, b):
    predictions = a * x + b
    a_gradient = -2 / len(x) * np.sum(x * (y - predictions))
    b_gradient = -2 / len(y) * np.sum(y - predictions)
    return np.array([a_gradient, b_gradient])
```

```python
def perform_gradient_descent(x, y, a, b, learning_rate):
    a_gradient = -2 / len(x) * np.sum(x * (y - (a * x + b)))
    b_gradient = -2 / len(y) * np.sum(y - (a * x + b))
    new_a = a - a_gradient * learning_rate
    new_b = b - b_gradient * learning_rate
    return (new_a, new_b)
```

```python
model_a, model_b = -5, 10 # Start points; can be anywhere
alpha = 0.01 # Learning rate
for step in range(1001):
    model_a, model_b = perform_gradient_descent(x, y, model_a, model_b, alpha)
    if step % 100 == 0:
        error = compute_cost(x, y, model_a, model_b)
        print("Step {}: a = {}, b = {}, J = {}".format(step, model_a, model_b, error))
print("Final line: {} * x + {}".format(model_a, model_b))
```

```python
# cost function for logistic regression

def compute_cost_log(x, predictions):
    x = x.reshape(-1, 1)
    # predictions = classification_model.predict_proba(x.reshape(-1, 1))
    distances = y * np.log(predictions) + (1 - y) * np.log(1 - predictions)
    result = -1 / len(x) * np.mean(distances)
    return result
```

```python
def compute_gradient_bce(y, y_pred_prob, x):
    error = y_pred_prob - y
    grad_w = (1 / len(y)) * np.dot(x.T, error)
    grad_b = (1 / len(y)) * np.sum(error)
    return grad_w, grad_b
```

```python
# Gradient descent parameters
alpha = 0.01  # Learning rate

# Perform gradient descent step
def perform_gradient_descent(x, predictions, weights, bias, alpha):
    N = len(predictions)
    z = np.dot(x, weights) + bias
    y_pred_prob = sigmoid(z)
    error = y_pred_prob - predictions
    
    grad_w = (1 / N) * np.dot(x.T, error) 
    grad_b = (1 / N) * np.sum(error)  
    
    weights -= alpha * grad_w  # Update weights
    bias -= alpha * grad_b  # Update bias
    
    return weights, bias


# Run gradient descent
for step in range(1001):
    weights, bias = perform_gradient_descent(x, y_pred_prob, weights, bias, alpha)
    if step % 100 == 0:
        cost = compute_cost(x, y_pred_prob, weights, bias)
        print(f"Step {step}: weights = {weights}, bias = {bias}, cost = {cost:.4f}")

print(f"Final model: weights = {weights}, bias = {bias}")

# It does not work...
```
