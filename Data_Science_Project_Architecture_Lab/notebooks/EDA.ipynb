```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# EDA

```python
asthma_data = pd.read_csv("../data/asthma_data_reduced.csv")
asthma_data
```

## Diagnosis

```python
diagnosis_mapping = {0: 'No', 1: 'Yes'}
diagnosis_counts = asthma_data.diagnosis.map(diagnosis_mapping).value_counts()

diagnosis_counts.plot(kind = 'bar')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.title('Asthma Diagnosis')
plt.xticks(rotation = 0)
plt.show()
```

```python
asthma_data.diagnosis.value_counts()
```

Asthma is diagnosed in 5.2% of the studied individuals.

## Age

```python
plt.hist(asthma_data.age[asthma_data.diagnosis == 1], bins = 20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

Asthma is observed in all studied age groups, but it is more frequent in the youngest and elderest persons.

## Sex

```python
sex_mapping = {0: 'Males', 1: 'Females'}
sex_counts = asthma_data.gender[asthma_data.diagnosis == 1].map(sex_mapping).value_counts()

sex_counts.plot(kind = 'bar', color = ['pink', 'blue'])
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Sex Distribution')
plt.xticks(rotation = 0)
plt.show()
```

Asthma is equally observed in males and females.

```python
ethnicity_mapping = {0: 'Caucasian', 1: 'African American', 2: 'Asian', 3: 'Other'}
ethnicity_counts = asthma_data.ethnicity[asthma_data.diagnosis == 1].map(ethnicity_mapping).value_counts()

ethnicity_counts.plot(kind = 'bar')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.title('Ethnicity Distribution')
plt.xticks(rotation = 0)
plt.show()
```

```python
education_mapping = {0: 'None', 1: 'High School', 2: 'Bachelor\'s', 3: 'Higher'}
education_counts = asthma_data.education_level[asthma_data.diagnosis == 1].map(education_mapping).value_counts()

education_counts.plot(kind = 'bar')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.title('Education Level Distribution')
plt.xticks(rotation = 0)
plt.show()
```

# Correlations among the variables

```python
correlation_matrix = asthma_data.corr("spearman")
plt.figure(figsize=(40, 36))
sns.heatmap(correlation_matrix, annot = True, annot_kws={"size": 18, "va": "center"}, fmt = ".2f", cbar = True, cmap = "coolwarm")
plt.show()
```

Weak correlations between variables. The correlation matrix is the same when using Pearson or Kendall correlation coefficients.
